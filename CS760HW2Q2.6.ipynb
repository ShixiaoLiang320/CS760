```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data (assuming space-separated values)
d1 = pd.read_csv('D1.txt', sep=' ', header=None)
d2 = pd.read_csv('D2.txt', sep=' ', header=None)

# Scatter plot of D1
plt.scatter(d1[0], d1[1])
plt.title('Scatter plot of D1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Scatter plot of D2
plt.scatter(d2[0], d2[1])
plt.title('Scatter plot of D2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np

def plot_decision_boundary(clf, X, y, ax):
    plot_step = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3)

# Fit and visualize decision tree for D1
clf1 = DecisionTreeClassifier()
clf1.fit(d1.iloc[:, :-1], d1.iloc[:, -1])
fig, ax = plt.subplots()
plot_decision_boundary(clf1, d1.iloc[:, :-1].values, d1.iloc[:, -1].values, ax)
ax.scatter(d1[0], d1[1], c=d1.iloc[:, -1])
plt.savefig('Decision Boundary for D1')
plt.title('Decision Boundary for D1')
plt.show()

# Repeat for D2
clf2 = DecisionTreeClassifier()
clf2.fit(d2.iloc[:, :-1], d2.iloc[:, -1])
fig, ax = plt.subplots()
plot_decision_boundary(clf2, d2.iloc[:, :-1].values, d2.iloc[:, -1].values, ax)
ax.scatter(d2[0], d2[1], c=d2.iloc[:, -1])
plt.savefig('Decision Boundary for D2')
plt.title('Decision Boundary for D2')
plt.show()
```
