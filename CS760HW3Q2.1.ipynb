```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('D2z.txt')
X_train = data[:, :2]  # Features
y_train = data[:, 2]   # Labels

x = np.arange(-2, 2.1, 0.1)
y = np.arange(-2, 2.1, 0.1)
xx, yy = np.meshgrid(x, y)
grid_points = np.c_[xx.ravel(), yy.ravel()]

def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2, axis=1))

predictions = []
for point in grid_points:
    distances = euclidean_distance(X_train, point)
    nearest = np.argmin(distances)
    predictions.append(y_train[nearest])

predictions = np.array(predictions).reshape(xx.shape)

plt.contourf(xx, yy, predictions, alpha=0.3, cmap='coolwarm')
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='coolwarm', edgecolors='k')
plt.title('1-NN Prediction on 2D Grid')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.savefig('1-NN Prediction on 2D Grid')
plt.show()
```
