```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = [1, -1]  # Mean 
cov_matrix = [[1, 0], [0, 1]]  # Covariance 

np.random.seed(0)  # Set a seed
samples = np.random.multivariate_normal(mean, cov_matrix, 100)

# Extract the x and y coordinates from the samples
x = samples[:, 0]
y = samples[:, 1]

# Create a scatter plot
plt.scatter(x, y, marker='o', c='b', label='Gaussian Data')
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the plot
plt.title('Scatter Plot of 100 Samples from a 2D Gaussian Distribution')
plt.grid(True)
plt.show()
```

```python
mean_1 = [5, 0] 
cov_matrix_1 = [[1, 0.25], [0.25, 1]]  
mean_2 = [-5, 0] 
cov_matrix_2 = [[1, -0.25], [-0.25, 1]] 

mean = 0.3 * mean_1 + 0.7 * mean_2
cov_matrix = 0.3*cov_matrix_1+0.7*cov_matrix_2

np.random.seed(0)  
samples = np.random.multivariate_normal(mean, cov_matrix, 100)

x = samples[:, 0]
y = samples[:, 1]

plt.scatter(x, y, marker='o', c='b', label='Gaussian Data')
plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.title('Scatter Plot of 100 Samples from a 2D Gaussian Distribution')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean_1 = np.array([5, 0])
cov_matrix_1 = np.array([[1, 0.25], [0.25, 1]])

mean_2 = np.array([-5, 0])
cov_matrix_2 = np.array([[1, -0.25], [-0.25, 1]])

weight_1 = 0.3
weight_2 = 0.7

combined_mean = weight_1 * mean_1 + weight_2 * mean_2
combined_cov_matrix = (weight_1**2) * cov_matrix_1 + (weight_2**2) * cov_matrix_2

np.random.seed(0)
samples = np.random.multivariate_normal(combined_mean, combined_cov_matrix, 100)

x = samples[:, 0]
y = samples[:, 1]

plt.scatter(x, y, marker='o', c='b', label='Gaussian Data')
plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.title('Scatter Plot of 100 Samples from a Combined 2D Gaussian Distribution')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0.01, 10, 400)  # Choose a range of x values, avoiding x <= 0 for log2

# Calculate the corresponding y values for ln(x) and log2(x)
y_ln = np.log(x)
y_log2 = np.log2(x)

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the figure size
plt.plot(x, y_ln, label='ln(x)', color='blue')
plt.plot(x, y_log2, label='log2(x)', color='red')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of ln(x) and log2(x)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0.01, 10, 400)  # Choose a range of x values, avoiding x <= 0 for log2

# Calculate the corresponding y values for ln(x) and log2(x)
#y_ln = np.log(x)
y_log2 = np.log2(x)
y_log2log2 = np.log2(np.log2(x))

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the figure size
plt.plot(x, y_log2, label = 'log2(x)', color='blue')
plt.plot(x, y_log2log2,label = 'log2log2(x)', color='red')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of log2log2(x) and log2(x)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of n values
n = np.arange(0, 10, 1)  # You can adjust the range as needed

# Calculate the corresponding values for n! and 2^n
factorial_n = [np.math.factorial(i) for i in n]
two_power_n = 2**n

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the figure size
plt.plot(n, factorial_n, label='n!', marker='o', color='blue')
plt.plot(n, two_power_n, label='2^n', marker='x', color='red')

# Add labels and legend
plt.xlabel('n')
plt.ylabel('Value')
plt.title('Comparison of n! and 2^n')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
